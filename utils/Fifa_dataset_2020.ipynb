```
import pandas as pd
import numpy as np
import json
```

```
df = pd.read_json('../data/json/players_stats_raw.json', encoding='utf-8-sig')
len(df.index)
```

```
df = df.drop_duplicates(subset='name')
df.head()
```

```
################# Data Cleaning #################
#################################################

# Split string into list
df['player_traits'] = df['player_traits'].apply(lambda trait_str: trait_str.split(','))

# Split string into list
df['player_hashtags'] = df['player_hashtags'].apply(lambda specials_str: specials_str.split(','))

df.head()
```

```
cols = df.columns[df.columns.isin(['name','Value', 'Wage', 'Release Clause'])]
print(cols)
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
df
```

```
# Export modified dataframe to json
df.to_json('../data/json/players_stats_modified.json', orient='records', force_ascii=False)

# Export modified dataframe to csv
df.to_csv('../data/csv/players_stats.csv', index=False, encoding=False)
```
